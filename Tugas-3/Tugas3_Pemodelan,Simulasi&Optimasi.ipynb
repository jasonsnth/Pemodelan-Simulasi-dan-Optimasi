```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
#DIKETAHUI
lambda_per_10min = 5
lambda_per_hour = lambda_per_10min * 6
```

```python
# Simulasi 1 Jam
one_hour = np.random.poisson(lam=lambda_per_hour)

one_hour
```

```python
# Simulasi 100 hari
np.random.seed(50)
one_hundred_days = np.random.poisson(lam=lambda_per_hour, size=100)

one_hundred_days
```

```python
# Histogram (Simulasi 100 hari)
plt.figure()
plt.hist(one_hundred_days, bins=10)
plt.title("Histogram Jumlah Pelanggan per Jam (Simulasi 100 Hari)")
plt.xlabel("Jumlah pelanggan")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# Hitung rata-rata dan variansi
mean_val = np.mean(one_hundred_days)
var_val = np.var(one_hundred_days)

print("Rata-rata:", mean_val)
print("Variansi:", var_val)
```

```python
# Probabilitas melebihi kapasitas 40 pelanggan
prob_sim = np.mean(one_hundred_days > 40)
prob_teo = 1 - poisson.cdf(40, lambda_per_hour)

print("Probabilitas simulasi:", prob_sim)
print("Probabilitas teoritis:", prob_teo)
```
